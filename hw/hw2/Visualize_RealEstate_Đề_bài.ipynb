2. Cho đoạn code minh họa Biểu đồ thể hiện số bài đăng theo tháng của năm 2020. Hãy thực hiện các yêu cầu cho các câu hỏi dưới đây:

```python
df1 = newdf.groupby('month').size().reset_index(name='number of records')
n = df1['month'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df1['month'], df1['number of records'], color=c, width=.5)
for i, val in enumerate(df1['number of records'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df1['month'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện số bài đăng theo tháng của năm 2020", fontsize=22)
plt.ylabel('number of records')
plt.show()
```

* Biểu đồ thể hiện giá nhà đất trung bình theo tháng (VND/m2)

```python
#Code ở đây
newdf['price_per_m2'] = newdf['price'] / newdf['square']

df2 = newdf.groupby('month')['price_per_m2'].mean().reset_index(name='avg_price_per_m2')
n = df2['month'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df2['month'], df2['avg_price_per_m2'], color=c, width=.5)
for i, val in enumerate(df2['avg_price_per_m2'].values):
    plt.text(i, val, round(val, 2), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df2['month'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện giá nhà đất trung bình theo tháng (VND/m²)", fontsize=22)
plt.ylabel('Giá trung bình (VND/m²)')
plt.xlabel('Tháng')
plt.show()
```

* Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)

```python
#Gọi ý: sns.displot()
plt.figure(figsize=(13,10), dpi= 80)
sns.distplot(newdf.loc[newdf['month'] == '12', "price_per_m2"], color="pink", label="Thang 12", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(newdf.loc[newdf['month'] == '04', "price_per_m2"], color="skyblue", label="Thang 4", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
plt.xlim(0, 120)
plt.title('Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)', fontsize=2)
plt.legend()
plt.show()
```

* Số lượng bất động sản phân theo số lượng bedrooms

```python
#Code ở đây
df_bed = newdf.groupby('bedrooms').size().reset_index(name='number_of_properties')
df_bed = df_bed.sort_values('bedrooms')
n = df_bed['bedrooms'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df_bed['bedrooms'], df_bed['number_of_properties'], color=c, width=.5)
for x, val in zip(df_bed['bedrooms'], df_bed['number_of_properties']):
    plt.text(x, val, f"{int(val)}",
             ha='center', va='bottom',
             fontdict={'fontweight':500, 'size':12})

plt.xticks(df_bed['bedrooms'])
plt.title("Số lượng bất động sản phân theo số lượng bedrooms", fontsize=22)
plt.xlabel("Số lượng phòng ngủ (bedrooms)", fontsize=14)
plt.ylabel("Số lượng bất động sản", fontsize=14)
plt.tight_layout()
plt.show()
```

* Giá bất động sản trung bình theo số lượng bedrooms

```python
# Code ở đây
df_bed = newdf.groupby('bedrooms')['price_per_m2'].mean().reset_index(name='avg_price')

n = df_bed['bedrooms'].nunique() + 1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi=80)
plt.bar(df_bed['bedrooms'], df_bed['avg_price'], color=c, width=.5)

for i, val in enumerate(df_bed['avg_price'].values):
    plt.text(df_bed['bedrooms'][i], val + 0.5, f"{val:.2f}",
             ha='center', va='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticks(df_bed['bedrooms'])
plt.gca().set_xticklabels(df_bed['bedrooms'], rotation=60, ha='right')

plt.title("Giá bất động sản trung bình theo số lượng bedrooms", fontsize=22)
plt.ylim(20,40)
plt.ylabel('Giá trung bình (VND/m2)')
plt.xlabel('Số lượng bedrooms')
plt.show()
```

* Giá nhà đất trung bình theo quận (VND/m2)

```python
# Code ở đây
df_district = newdf.groupby('district')['price_per_m2'].mean().reset_index(name='avg_price')
df_district = df_district.sort_values(by = 'avg_price')
n = df_district['district'].nunique() + 1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi=80)
plt.bar(df_district['district'], df_district['avg_price'], color=c, width=.5)

for i, val in enumerate(df_district['avg_price'].values):
    plt.text(i, val, f"{val:.2f}",
             ha='center', va='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticks(df_district['district'])
plt.gca().set_xticklabels(df_district['district'], rotation=60, ha='right')

plt.title("Giá nhà đất trung bình theo quận (VND/m2)", fontsize=22)
plt.ylabel('Giá trung bình (VND/m2)')
plt.xlabel('Quận')
plt.show()
```

* Giá nhà đất trung bình theo các phường quận Hoàn Kiếm (VND/m2)

```python
# Code ở đây
df_hoankiem = newdf[newdf['district'] == 'hoan kiem']

df_ward = df_hoankiem.groupby('ward')['price_per_m2'].mean().reset_index(name='avg_price')
df_ward = df_ward.sort_values(by='avg_price')

n = df_ward['ward'].nunique() + 1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi=80)
plt.bar(range(len(df_ward)), df_ward['avg_price'], color=c, width=0.5)

for i, val in enumerate(df_ward['avg_price']):
    plt.text(i, val, f"{val:.2f}", ha='center', va='bottom', fontdict={'fontweight':500, 'size':12})

plt.xticks(range(len(df_ward)), df_ward['ward'], rotation=60, ha='right')
plt.title("Giá nhà đất trung bình theo các phường quận Hoàn Kiếm (VND/m2)", fontsize=22)
plt.ylabel('Giá trung bình (VND/m2)')
plt.xlabel('Phường')
plt.show()
```

* Giá nhà đất trung bình theo nhà đầu tư (VND/m2)

```python
# Code ở đây
df_investor = newdf.groupby('investor')['price_per_m2'].mean().reset_index(name='avg_price')
df_investor = df_investor.sort_values(by='avg_price',ascending= False).head(10)
n = df_investor['investor'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi=80)
plt.bar(range(len(df_investor)), df_investor['avg_price'], color=c, width=0.5)

for i, val in enumerate(df_investor['avg_price']):
    plt.text(i, val, f"{val:.2f}", ha='center', va='bottom', fontdict={'fontweight':500, 'size':12})

plt.xticks(range(len(df_investor)), df_investor['investor'], rotation=60, ha='right')
plt.title("Giá nhà đất trung bình theo nhà đầu tư (VND/m2)", fontsize=22)
plt.ylim(40,65)
plt.ylabel('Giá trung bình (VND/m2)')
plt.xlabel('Nhà đầu tư')
plt.show()
```

* Giá nhà đất trung bình theo dự án (VND/m2)

```python
# Code ở đây
df_project = newdf.groupby('project')['price_per_m2'].mean().reset_index(name='avg_price')
df_project = df_project.sort_values(by='avg_price',ascending= False).head(10)
n = df_project['project'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi=80)
plt.bar(range(len(df_project)), df_project['avg_price'], color=c, width=0.5)

for i, val in enumerate(df_project['avg_price']):
    plt.text(i, val, f"{val:.2f}", ha='center', va='bottom', fontdict={'fontweight':500, 'size':12})

plt.xticks(range(len(df_project)), df_project['project'], rotation=60, ha='right')
plt.ylim(50,80)
plt.title("Giá nhà đất trung bình theo dự án (VND/m2))", fontsize=22)
plt.ylabel('Giá trung bình (VND/m2)')
plt.show()
```

* Giá bất động sản phân bố theo top 10 nhà đầu tư triệu /m2

```python
# Code ở đây
df_investor = newdf[newdf['investor'].notnull()]
df_avg = df_investor.groupby('investor')['price_per_m2'].mean().reset_index(name='avg_price')

top10_investors = df_avg.sort_values(by='avg_price', ascending=False).head(10)['investor']
df_top10 = df_investor[df_investor['investor'].isin(top10_investors)]

plt.figure(figsize=(16,10), dpi=80)
sns.boxplot(x='investor', y='price_per_m2', data=df_top10)
plt.title("Giá bất động sản phân bố theo top 10 nhà đầu tư (triệu/m²)", fontsize=22)
plt.xlabel("Nhà đầu tư")
plt.ylabel("Giá (triệu/m²)")
plt.xticks(rotation=45, ha='right')
plt.show()
```

* Giá bất động sản phân bố theo dự án

```python
# Code ở đây
df_project = newdf[newdf['project'].notnull()]
df_avg = df_project.groupby('project')['price_per_m2'].mean().reset_index(name='avg_price')
top10_project = df_avg.sort_values(by='avg_price',ascending= False).head(10)['project']
df_top10_pr = df_project[df_project['project'].isin(top10_project)]

plt.figure(figsize=(16,10), dpi=80)
sns.boxplot(x='project', y='price_per_m2', data=df_top10_pr)
plt.title("Giá bất động sản phân bố theo dự án", fontsize=22)
plt.xlabel("Dự án")
plt.ylabel("Giá")
plt.xticks(rotation=45, ha='right')
plt.show()
```

* Giá nhà đất trung bình theo hướng nhà (VND/m2)

```python
# Code ở đây
df_direction = newdf.groupby('direction')['price_per_m2'].mean().reset_index(name='avg_price')
df_direction = df_direction.sort_values(by='avg_price')
n = df_direction['direction'].unique().__len__() + 1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi=80)
plt.bar(range(len(df_direction)), df_direction['avg_price'], color=c, width=0.5)

for i, val in enumerate(df_direction['avg_price']):
    plt.text(i, val, f"{val:.2f}", ha='center', va='bottom', fontdict={'fontweight':500, 'size':12})

plt.xticks(range(len(df_direction)), df_direction['direction'], rotation=60, ha='right')
plt.ylim(20,40)
plt.title("Giá nhà đất trung bình theo hướng nhà (VND/m2)", fontsize=22)
plt.ylabel('Giá trung bình (VND/m2)')
plt.xlabel('Hướng nhà')
plt.show()
```
